```python
import pandas as pd

df = pd.read_csv('sales.csv')
display(df.head())
```

```python
#T TEST TWO-SAMPLE

from scipy.stats import ttest_ind

member_spending = df[df['Customer type']== 'Member']['Total']
normal_spending = df[df['Customer type'] == 'Normal']['Total']

t_stat, p_value = ttest_ind(member_spending, normal_spending)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

# **So here the null hypothesis : there is no significant difference between average spending of Member and Normal customer.**

# null hypothesis is accepted. As p-value > 0.05

```python
#ANOVA F-TEST

from scipy.stats import f_oneway

city_a = df[df['City'] == 'Vizag']['gross income']
city_b = df[df['City'] == 'Banglore']['gross income']
city_c = df[df['City'] == 'Hyderabad']['gross income']

f_stats,p_value = f_oneway(city_a,city_b,city_c)

print(f'f-statistics: {f_stats}')
print(f'p-value: {p_value}')
```

# **NULL HYPOTHESIS is: there is no significant difference between average gross income across the 3 cities.**


now here p-value > 0.05 so null hypothesis is accepted.
i.e even thought there is some difference in gross income (i.e vizag has highest gross income) it's statistically not significant.

```python
# PEARSON CORRELATION
from scipy.stats import pearsonr

correlation, p_value = pearsonr(df['Rating'],df['Total'])

print(f'correlation: {correlation}')
print(f'p-value: {p_value}')

if p_value < 0.05:
    print("Statistically significant correlation between Total and Rating.")
else:
    print("No statistically significant correlation between Total and Rating.")
```

# null hypothesis : there is no significant difference between total money spent and the ratings given .
p-value > 0.05 , so null hypothesis is accepted.

```python
#CHI-SQUARED TEST

import pandas as pd
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['Gender'], df['Product line'])
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

print(f'Chi-squared statistic: {chi2_stat:.4f}')
print(f'p-value: {p_val:.4f}')

if p_val < 0.05:
    print("Product line preference is dependent on Gender.")
else:
    print("Product line preference is independent of Gender.")
```
