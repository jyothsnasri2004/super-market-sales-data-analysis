```python
from prophet import Prophet
import pandas as pd

df_branch = df[df['Branch'] == 'A'][['date', 'Total']].rename(columns={'date': 'ds', 'Total': 'y'})

model = Prophet()
model.fit(df_branch)

future = model.make_future_dataframe(periods=3, freq='Y')  # Forecasting next 3 years
forecast = model.predict(future)

model.plot(forecast)
```

```python
product_lines = df['Product line'].unique()
future_demand = {}

for product in product_lines:
    df_prod = df[df['Product line'] == product]
    df_grouped = df_prod.groupby('date')['Quantity'].sum().reset_index()
    df_grouped = df_grouped.rename(columns={'date': 'ds', 'Quantity': 'y'})

    model = Prophet()
    model.fit(df_grouped)

    future = model.make_future_dataframe(periods=3, freq='M')  # next 3 months
    forecast = model.predict(future)

    # Filter only forecasted dates
    forecast_future = forecast[forecast['ds'] > df_grouped['ds'].max()]
    total_predicted_quantity = forecast_future['yhat'].sum()

    future_demand[product] = total_predicted_quantity

    # Sort by predicted demand (descending)
sorted_demand = sorted(future_demand.items(), key=lambda x: x[1], reverse=True)

# Display Top Products
print("Predicted Demand (Top Product Lines for Next Quarter):")
for product, qty in sorted_demand:
    print(f"{product}: {qty:.2f} units")

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Load and prepare data
df = pd.read_csv("sales.csv")
df = df.drop(columns=['Invoice ID', 'Date', 'Time'], errors='ignore')

X = df.drop(columns=['Total'])
y = df['Total']

# Encode categorical variables
categorical_cols = X.select_dtypes(include='object').columns
preprocessor = ColumnTransformer(
    transformers=[('cat', OneHotEncoder(drop='first'), categorical_cols)],
    remainder='passthrough'
)

X_encoded = preprocessor.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict & evaluate
y_pred = model.predict(X_test)

df['Predicted_Total'] = model.predict(preprocessor.transform(X))
sales_by_type = df.groupby('Customer type')['Predicted_Total'].sum()
print(sales_by_type)
```
